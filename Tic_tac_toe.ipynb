```python
#from IPython.core.debugger import set_trace
#Global variables
game_still_on=True
winner=None
current_player=' X '

#board
board=[" - "," - "," - ",
       " - "," - "," - ",
       " - "," - "," - "]

def display_board():
    #displays board to bothe the players along with numbers.
    print("\n")
    print(board[0] + " | " + board[1] + " | " + board[2] + "     1 | 2 | 3")
    print(" ---"+'|'+" --- "+"|"+" ---")
    print(board[3] + " | " + board[4] + " | " + board[5] + "     4 | 5 | 6")
    print(" ---"+'|'+" --- "+"|"+" ---")
    print(board[6] + " | " + board[7] + " | " + board[8] + "     7 | 8 | 9")
    print("\n")
    
def handle_turn(player):
    #Fills value in the board according to the user 
    #Dssplays which player's turn is there now.
    print("\n" + player +"'s turn:")
    value=(input("Enter the choice:"))
    #valid:Flag to get the correct input from the user.
    valid=True
    while(valid):
        
        while(value not in ['1','2','3','4','5','6','7','8','9']):
            value=(input("Invalid input.\nchoose the correct choice:"))
        value=int((value))-1
        #Check whether the box is filled or not.
        if(board[value]==' - '):
            valid=False    
        else:
            #To avoid the over-riding of the X and O on the board.
            print("You can't go there.")
    #Fill the respect value in the box on the board.        
    board[value]=player
    display_board()  

def check_if_game_over():
    check_for_winner()
    check_if_tie() 

def check_for_winner():
    global winner
    #check for rows
    row=check_rows()
    #print(row)
    
    #check for columns
    column=check_columns()
    
    #check for diagonal
    diagonal=check_diagonals()
    
    #Get the winner:
    if row:
        winner=row
    elif column:
        winner=column
    elif diagonal:
        winner=diagonal
    else:
        winner=None

def check_rows():
    global game_still_on
    
    row_1=board[0]==board[1]==board[2]!=" - "
    row_2=board[3]==board[4]==board[5]!=" - "
    row_3=board[6]==board[7]==board[8]!=" - "
    
    #Stop the game if winner is found
    if row_1 or row_2 or row_3:
        game_still_on=False
    if row_1:
        return board[0]
    elif row_2:
        return board[3]
    elif row_3:
        return board[6]
    else:
        return None

def check_columns():
    global game_still_on
    
    column_1=board[0]==board[3]==board[6]!=" - "
    column_2=board[1]==board[4]==board[7]!=" - "
    column_3=board[2]==board[5]==board[8]!=" - "
    
    #Stop the game if winner is found
    if column_1 or column_2 or column_3:
        game_still_on=False
    if column_1:
        return board[0]
    elif column_2:
        return board[1]
    elif column_3:
        return board[2]
    else:
        return None

def check_diagonals():
    global game_still_on
    
    diagonal_1=board[0]==board[4]==board[8]!=" - "
    diagonal_2=board[2]==board[4]==board[6]!=" - "
    
    #Stop the game if winner is found
    if diagonal_1 or diagonal_2:
        game_still_on=False
    if diagonal_1:
        print(board[8])
        return board[8]
    elif diagonal_2:
        return board[6]
    else:
        return None

def check_if_tie():
    #If all the boxes filled in the board then stop the game and declare tie.
    global game_still_on
    if( ' - ' not in board):
        game_still_on=False

def flip_player():
    #Gives chance to both the players alternatively. 
    global current_player
    if current_player==" X ":
        current_player=" O "
    elif current_player==" O ":
        current_player=" X "

def play_game():
    
    display_board()
    
    while game_still_on: 
        #Handle single turn of each player
        handle_turn(current_player)
        #Check whether game is over or not
        check_if_game_over()
        #Switch chance to other player
        flip_player()
        
     #Game ends:
    if winner==" X " or winner==" O ":
        print('Game ends')
        print(winner + "won.")
    elif winner==None:
        print("Match tie.")
```
